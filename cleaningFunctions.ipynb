```
from pyspark.sql.functions import col,when,create_map,lit
from pyspark.sql import Column
from itertools import chain

def isDayOfWeek(column,data,args):  # args: empty list []
    return data.filter(col(column).between(1,7))
    
def isDayOfMonth(column,data,args):  # args: empty list []
    return data.filter(col(column).between(1,31))
    
def isMonth(column,data,args):  # args: empty list []
    return data.filter(col(column).between(1,12))
    
def isYear(column,data,args):  # args: empty list []
    return data.filter(col(column).between(1900,2050))

def removeIfNull(column,data,args):  # args: empty list []
    return data.filter(col(column).isNotNull())
    
def isBetween(column,data,args):  # args: list of values [lower, higher]
    return data.filter(col(column).between(args[0],args[1]))

def isLargerThan(column,data,args):  # args: single value (passed inside list), e.g. [0]
    return data.filter(col(column) > args[0])

def isLargerThanCol(column,data,args):  # args: single column target-name (passed inside list), e.g. ["DepTime"]
    return data.filter(col(column) > col(args[0]))

def isLessThan(column,data,args):  # args: single value (passed inside list), e.g. [0]
    return data.filter(col(column) < args[0])

def isLessThanCol(column,data,args):  # args: single column target-name (passed inside list), e.g. ["DepTime"]
    return data.filter(col(column) < col(args[0]))

def isDifferentFrom(column,data,args):  # args: single column target-name (passed inside list), e.g. ["Origin"]
    return data.filter(col(column) != col(args[0]))  
    
def changeUsingMap(column,data,args):  # args: single dictionary (passed inside list), e.g. [{"YES":"Y","NO":"N"}]
    return data.replace(args[0],subset=column)
    
func = {
    "isDayOfWeek": isDayOfWeek,
    "isDayOfMonth": isDayOfMonth,
    "isMonth": isMonth,
    "isYear": isYear,
    "removeIfNull": removeIfNull,
    "isBetween": isBetween,
    "isLargerThan": isLargerThan,
    "isLargerThanCol": isLargerThanCol,
    "isLessThan": isLessThan,
    "isLessThanCol": isLessThanCol,
    "isDifferentFrom": isDifferentFrom,
    "changeUsingMap": changeUsingMap
}
```
